 ## 2-1.axis (열/행을 드랍)

axis=0은 행을 드랍합니다.

```
df.dropna(axis=0)
```

axis=1은 열을 드랍합니다.

```
df.dropna(axis=1)
```

# 2-2. how 옵션 -'any': 한개라도 있는 경우 드랍, 'all'은 모두 NaN인 경우 드랍

```
df.dropna()
```

```
df.dropna(axis=0, how='any')
```

```
df.dropna(axis=0, how='all')
```

## Barplot 

```
x = ['Math', 'Programming', 'Data Science', 'Art', 'English', 'Phtsics']
y = [66, 80, 60, 50, 80, 10]

plt.bar(x, y, align='center', alpha=0.7, color='red')
plt.xticks(x)
plt.ylabel('Number of Students')
plt.title('Subjects')

plt.show()
```

```
x = ['Math', 'Programming', 'Data Science', 'Art', 'English', 'Phtsics']
y = [66, 80, 60, 50, 80, 10]

# plt.figure(figsize=(6, 3))
plt.bar(x, y, align='center', alpha=0.7, color='red')
plt.xticks(x, rotation=90)
plt.ylabel('Number of Students')
plt.title('Subjects')

plt.show()
```

```
x = ['Math', 'Programming', 'Data Science', 'Art', 'English', 'Phtsics']
y = [66, 80, 60, 50, 80, 10]

# plt.figure(figsize=(6, 3))
plt.barh(x, y, align='center', alpha=0.7, color='green')
plt.yticks(x)
plt.ylabel('Number of Students')
plt.title('Subjects')

plt.show()
```

```
x_label = ['Math', 'Programming', 'Data Science', 'Art', 'English', 'Phtsics']
x = np.arange(len(x_label))
y1 = [66, 80, 60, 50, 80, 10]
y2 = [55, 90, 40, 60, 70, 20]

# 
width = 0.35

# subplots 
fig, axes = plt.subplots()

#  
axes.bar(x - width/2, y1, width, align='center', alpha=0.5)
axes.bar(x + width/2, y2, width, align='center', alpha=0.8)

# xtick 
plt.xticks(x)
axes.set_xticklabels(x_label)
plt.ylabel('Number of Students')
plt.title('Subjects')

plt.legend(['john', 'peter'])


plt.show()

```
